**form the above observation here are few points which will help us to get efficient result**

**Purchases Table:
The purchases table contains actual purchase data, including purchase dates, products (brands) purchased by stores from vendors, the amount paid (in dollars), and the quantity purchased.**

**Purchase Price Reference:
The PurchasePrice column is related to the purchase_prices table, which provides brand-wise standard and actual purchase prices. The combination of VendorNumber and Brand is assumed to be unique in this table.**

**Vendor Invoice Table:
The vendor_invoice table summarizes purchase activity, aggregating total quantity and dollar amounts per invoice, with an additional column for freight costs. The table is unique per VendorNumber and PONumber combination.**

**Sales Table:
The sales table captures actual sales transactions by stores, detailing the brands sold, the quantities sold, the sales prices, and the revenue generated.**

**As the data that we need for analysis is disributed in differnet tables, we need to create a summary table containing :** 

```python
"""
Purchase transactions made by vendors to stores

Sales transaction data from stores to customers

Freight costs associated with each vendor (from vendor_invoice) #freight means logistics cost

Actual product prices from vendors (from purchase_prices)

"""
```

```python
# freight_summary = pd.read_sql_query(''' select VendorNumber, sum(Freight) as FreightCost From vendor_invoice Group BY VendorNumber ''', conn)
# print(freight_summary.shape)
# freight_summary.head()

pd.read_sql_query(''' select VendorNumber, sum(Freight) as FreightCost From vendor_invoice Group BY VendorNumber ''', conn)
```

```python
pd.read_sql_query("""SELECT    
            p.VendorNumber, 
            p.VendorName, 
            p.Brand, 
            p.PurchasePrice, 
            pp.volume,
            pp.price as ActualPrice,
            Sum(p.Quantity) as TotalPurchaseQunatity, 
            sum(p.Dollars) TotalPurchaseDollars 
            From purchases p
            Join purchase_prices pp
            on p.Brand = pp.Brand
            where p.PurchasePrice > 0
            Group by 1, 2, 3
            ORDER BY TotalPurchaseDollars""", conn)
```

```python
pd.read_sql_query('''select 
        VendorNo,
        Brand,
        SUM(SalesDollars) as TotalSalesDollars,
        Sum(SalesQuantity) as TotalSalesQuantity,
        Sum(ExciseTax) as TotalExciseTax
        From sales
        Group by VendorNo, Brand''', conn)
```

```python
vendor_sales_summary = pd.read_sql_query("""WITH freight_summary AS (
        SELECT 
            VendorNumber, 
            SUM(Freight) AS FreightCost 
        FROM vendor_invoice 
        GROUP BY VendorNumber
    ),
    purchase_summary AS (
        SELECT    
            p.VendorNumber, 
            p.VendorName, 
            p.Brand, 
            p.Description,
            p.PurchasePrice, 
            pp.volume,
            pp.price AS ActualPrice,
            SUM(p.Quantity) AS TotalPurchaseQuantity, 
            SUM(p.Dollars) AS TotalPurchaseDollars 
        FROM purchases p
        JOIN purchase_prices pp
            ON p.Brand = pp.Brand
        WHERE p.PurchasePrice > 0
        GROUP BY p.VendorNumber, p.VendorName, p.Brand
    ),
    sales_summary AS (
        SELECT 
            VendorNo,
            Brand,
            SUM(SalesDollars) AS TotalSalesDollars,
            SUM(SalesQuantity) AS TotalSalesQuantity,
            SUM(SalesPrice) AS TotalSalesPrice,
            SUM(ExciseTax) AS TotalExciseTax
        FROM sales
        GROUP BY VendorNo, Brand
    )
    SELECT 
        ps.vendorNumber,
        ps.VendorName,
        ps.Brand, 
        ps.Description, -- Fixed column name
        ps.PurchasePrice,
        ps.ActualPrice,
        ps.Volume,
        ps.TotalPurchaseQuantity, -- Fixed typo
        ps.TotalPurchaseDollars,
        ss.TotalSalesQuantity,
        ss.TotalSalesPrice,
        ss.TotalSalesDollars,
        ss.TotalExciseTax,
        fs.FreightCost
    FROM purchase_summary ps
    LEFT JOIN sales_summary ss
        ON ps.vendorNumber = ss.VendorNo
        AND ps.Brand = ss.Brand
    LEFT JOIN freight_summary fs
        ON ps.vendorNumber = fs.VendorNumber
    ORDER BY ps.TotalPurchaseDollars DESC;""",conn)
```

```python
vendor_sales_summary 
```

**This query generates sales and  purchase summary according to vendor basis which is valuable for:**

The query involves heavy joins and aggregations on large datasets like sales, purchases, and vendor_invoice.

Storing pre-aggregated results (e.g. in a vendor_sales_summary table) avoids repeated expensive computations.

This summary table helps in analyzing sales, purchases, pricing, and freight at the vendor and brand level.

It provides future benefits for faster dashboarding and reporting by reducing query complexity at runtime.

Instead of running expensive joins and aggregations each time, dashboards can fetch data quickly from the summary table.

```python
vendor_sales_summary['volume'] = vendor_sales_summary['volume'].astype('float64') #to change the volume datatype to float from object.
vendor_sales_summary.fillna(0, inplace=True)                                      #to fill the missing or null values with zero
vendor_sales_summary['VendorName']  = vendor_sales_summary['VendorName'].str.strip() #to remove the white spaces from vendorName column.
```

```python
#added new column to calculate profit of seperate vendors
vendor_sales_summary['GrossProfit'] = vendor_sales_summary['TotalSalesDollars'] - vendor_sales_summary['TotalPurchaseDollars']
```

```python
# vendor_sales_summary.min()
vendor_sales_summary['GrossProfit'].min()
```

```python
vendor_sales_summary['ProfitMargin'] = (vendor_sales_summary['GrossProfit'] / vendor_sales_summary['TotalSalesDollars']) * 100
```

```python
vendor_sales_summary['StockTurnover'] = vendor_sales_summary['TotalSalesQuantity']/ vendor_sales_summary['TotalPurchaseQuantity']
```

```python
vendor_sales_summary['SalestoPurchaseRatio'] = vendor_sales_summary['TotalSalesDollars']/ vendor_sales_summary['TotalPurchaseDollars']
```

**It's always a good practice to store a final table into a new table. so we'll create a new table and put values into this**

```python
cursor = conn.cursor()

cursor.execute(""" CREATE TABLE Vendor_Final_summary (
    VendorNumber INT,
    VendorName VARCHAR(100),
    Brand INT,
    Description VARCHAR(100),
    PurchasePrice DECIMAL(15,2),
    ActualPrice DECIMAL(15,2),
    volume DECIMAL(15,2),
    TotalPurchaseQunatity INT,
    TotalPurchaseDollars DECIMAL(15,2),
    TotalSalesQuantity DECIMAL(15,2),
    TotalSalesPrice DECIMAL(15,2),
    TotalSalesDollars DECIMAL(15,2),
    TotalExciseTax DECIMAL(15,2),
    FreightCost DECIMAL(15,2),
    GrossProfit DECIMAL(15,2),
    ProfitMargin DECIMAL(15,2),
    StockTurnover DECIMAL(15,2),
    SalestoPurchaseRatio DECIMAL(15,2),
    primary key(VendorNumber, Brand)
);
""")
```

```python
#uploaded the final table into database
pd.read_sql('''select count(*) as total_row from Vendor_Final_summary''', conn)
```

```python
vendor_sales_summary.to_sql('Vendor_Final_summary', conn, if_exists= 'replace', index = False)
```
